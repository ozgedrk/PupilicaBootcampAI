## 120 Yllk Olimpiyat Tarihi Veri Seti
* Bu, Atina 1896'dan Rio 2016'ya kadar tm oyunlar da dahil olmak zere modern Olimpiyat Oyunlarna ilikin tarihsel bir veri kmesidir. 

* Veri seti ierisinde bulunan stunlarn isimleri ve aklamalar
    1. ID - Her sporcu iin benzersiz numara
    1. Name - Sporcunun ismi
    1. Sex - Cinsiyet: M(Erkek) - F(Kadn)
    1. Age - Ya
    1. Height - Boy (santimetre)
    1. Weight - Arlk (Kilogram)
    1. Team - Takm ismi
    1. NOC - Ulusal Olimpiyat Komitesi 3 harfli kod
    1. Games - Yl ve sezon
    1. Year - Yl
    1. Season - Sezon: Yaz yada K
    1. City - Dzenlenen ehir
    1. Sport - Spor
    1. Event - Etkinlik
    1. Medal - Madalya: Altn, Gm, Bronz ve Madalya Yok
