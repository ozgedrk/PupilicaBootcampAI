## Seaborn Kütüphanesi
* Seaborn, Matplotlib kütüphanesi tabanlı, istatiksel bir Python veri görselleştirme kütüphanesidir. 
* Seaborn kullanıcılara istatiksel görselleştirmeler yapmaları için yüksek seviyeli bir arayüz sunar.
* Seaborn kütüphanesi’nin işlevleri aşağıdaki gibidir.
    * Birden çok değişken arasındaki ilişkileri detaylı bir şekilde incelemek adına sunulan veri kümesi tabanlı bir uygulama programlama arayüzüdür.
    * Kategorik değişkenleri gözlemlemek ve üzerinde istatistiksel işlemler yapmak için destek sağlar.
    * Karmaşık görselleştirmeler oluşturmayı sağlayançoklu grafikleri destekler.
    * Alt kümeler arasında karşılaştırma yapmak için tek değişkenli ve çift değişkenli görselleştirmeler sunar.
    * Görselleştirmeyi daha anlaşılır hale getiren geniş renk skalası mevcuttur.
* Seaborn ve Matplotlib kütüphanesi arasındaki farklar
    * Matplotlib kişiselleştirilebilir fakat plotları daha anlaşılır, göze hitap eden şekle getirmek bir hayli zordur.
    * Seaborn sahip olduğu renk paleti, arayüzler ve grafik çeşitleri ile kullanıcıya farklı çözümler sunar.
    * Pandas Kütüphanesi ile çalışırken Seaborn kolaylık sağlar.

## 120 Yıllık Olimpiyat Tarihi Veri Seti
* 120 Yıllık Olimpiyat Tarihi veri seti tanıtımını ve temizlenmesini veri bilimi bölümünde yaptığımız için bu bölümde verinin içe aktarılması işlemini yapacağız.

```python
# kütüphaneleri import edelim
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# pythonda uyarıları kapatır
import warnings
warnings.filterwarnings("ignore")
```

```python
# veriyi içe aktar
veri = pd.read_csv("olimpiyatlar_temizlenmis.csv")
veri.head()
```

## Saçılım Grafiği

```python
# Saçılım grafiği çizmek için bir plot yapalım
def sacilimGrafik():
    sns.scatterplot(x='boy', y='kilo', data=veri)
    plt.title('Boy ve Kilo Dağılımı - Beyaz Izgara Tema')
    plt.show()
sacilimGrafik()    
```

```python
# seaborn stilleri
sns.set_style("white")
sacilimGrafik()
```

## Saçılım Grafiği - Kategorik

```python
sns.scatterplot(x='boy', y='kilo', hue='madalya', data=veri)
plt.title('Madalyaya Göre Boy ve Kilo Dağılımı')
plt.show()
```

<a id='4'></a>
## Saçılım Grafiği - Doğrusal Regresyon

```python
sns.regplot(x='boy', y='kilo', data=veri, marker="+", scatter_kws={'alpha':0.2})
plt.title('Boy ve Kilo Dağılımı')
plt.show()
```

<a id='5'></a>
## Renk Paletleri

```python
# Renk Paletleri
sns.scatterplot(x='boy', y='kilo', hue='madalya', data=veri, palette = "Set1")
plt.title('Madalyaya Göre Boy ve Kilo Dağılımı')
plt.show()
```

```python
sns.scatterplot(x='boy', y='kilo', hue='madalya', data=veri, palette = "rocket")
plt.title('Madalyaya Göre Boy ve Kilo Dağılımı')
plt.show()
```

<a id='6'></a>
## Çizgi Grafiği

```python
sns.lineplot(x='boy', y='kilo', data=veri)
plt.title('Boy ve Kilo')
plt.show()
```

<a id='7'></a>
## Çizgi Grafiği - Kategorik

```python
sns.lineplot(x='boy', y='kilo', hue='cinsiyet', data=veri)
plt.title('Cinsiyete göre Boy ve Kilo')
plt.show()
```

<a id='8'></a>
## Histogram

```python
sns.displot(veri, x = "kilo")
plt.ylabel("Frekans")
plt.title("Kilo Histogram")
plt.show()
```

<a id='9'></a>
## Histogram - Kategorik

```python
sns.displot(veri, x="kilo", hue="cinsiyet")
plt.ylabel("Frekans")
plt.title("Cinsiyete Göre Kilo Histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", col="cinsiyet", multiple="dodge")
plt.show()
```

<a id='10'></a>
## İki Boyutlu Histogram

```python
sns.displot(veri, x="kilo", y="boy", kind="kde")
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("Kilo - Boy Histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy",  hue="cinsiyet", kind="kde")
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("Cinsiyete göre Kilo - Boy Histogram")
plt.show()
```

```python
sns.kdeplot(data = veri, x="kilo", y="boy",  hue="cinsiyet", fill = True)
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("Cinsiyete göre Kilo - Boy Histogram")
plt.show()
```

## Çubuk Grafiği

```python
sns.barplot(x="madalya", y="boy", data=veri)
plt.title("Madalyalara göre boy grafikleri")
plt.show()
```

```python
sns.barplot(x="madalya", y="yas", data=veri, hue="cinsiyet")
plt.title("Madalyalara göre boy grafikleri")
plt.show()
```

```python
sns.catplot(x="madalya", y="yas",  hue="cinsiyet", col="sezon",
                data=veri, kind="bar",height=4, aspect=.7)
plt.show()
```

<a id='12'></a>
## Kutu Grafiği

```python
sns.boxplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy", hue="cinsiyet",data=veri, palette="Set2")
plt.show()
```

```python
veri_gecici = veri.loc[:,["yas","boy","kilo"]]
sns.boxplot(data=veri_gecici, orient="h", palette="Set2")
plt.show()
```

```python
sns.catplot(x="sezon", y="boy",hue="cinsiyet", col="madalya",data=veri, kind="box",height=4, aspect=.7);
```

<a id='13'></a>
## Sıcaklık Haritası

```python
sns.heatmap(veri.corr(), annot=True, linewidths=.5, fmt= '.1f')
plt.show()
```

<a id='14'></a>
## Keman Grafiği

```python
sns.violinplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", data=veri, hue="cinsiyet")
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", data=veri, hue="cinsiyet", split=True)
plt.show()
```

```python
sns.catplot(x="sezon", y="boy",
                hue="cinsiyet", col="madalya",
                data=veri, kind="violin", split=True,
                height=4, aspect=.7)
plt.show()
```

<a id='15'></a>
## Ortaklık Grafiği

```python
sns.jointplot(
    data=veri,
    x="kilo", y="boy", hue="sezon",
    kind="kde")
plt.show()
```

```python
g = sns.JointGrid(data=veri, x="kilo", y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

<a id='16'></a>
## Çift/Eş Grafiği

```python
sns.pairplot(veri)
plt.show()
```

```python
g = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```

<a id='17'></a>
## Sayma Grafiği

```python
sns.countplot(x="sehir", data=veri)
plt.xticks(rotation = 90)
plt.show()
```
