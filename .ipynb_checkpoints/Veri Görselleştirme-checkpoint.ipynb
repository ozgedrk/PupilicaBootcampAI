## Seaborn Ktphanesi
* Seaborn, Matplotlib ktphanesi tabanl, istatiksel bir Python veri grselletirme ktphanesidir. 
* Seaborn kullanclara istatiksel grselletirmeler yapmalar iin yksek seviyeli bir arayz sunar.
* Seaborn ktphanesinin ilevleri aadaki gibidir.
    * Birden ok deiken arasndaki ilikileri detayl bir ekilde incelemek adna sunulan veri kmesi tabanl bir uygulama programlama arayzdr.
    * Kategorik deikenleri gzlemlemek ve zerinde istatistiksel ilemler yapmak iin destek salar.
    * Karmak grselletirmeler oluturmay salayanoklu grafikleri destekler.
    * Alt kmeler arasnda karlatrma yapmak iin tek deikenli ve ift deikenli grselletirmeler sunar.
    * Grselletirmeyi daha anlalr hale getiren geni renk skalas mevcuttur.
* Seaborn ve Matplotlib ktphanesi arasndaki farklar
    * Matplotlib kiiselletirilebilir fakat plotlar daha anlalr, gze hitap eden ekle getirmek bir hayli zordur.
    * Seaborn sahip olduu renk paleti, arayzler ve grafik eitleri ile kullancya farkl zmler sunar.
    * Pandas Ktphanesi ile alrken Seaborn kolaylk salar.

## 120 Yllk Olimpiyat Tarihi Veri Seti
* 120 Yllk Olimpiyat Tarihi veri seti tantmn ve temizlenmesini veri bilimi blmnde yaptmz iin bu blmde verinin ie aktarlmas ilemini yapacaz.

```python
# ktphaneleri import edelim
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# pythonda uyarlar kapatr
import warnings
warnings.filterwarnings("ignore")
```

```python
# veriyi ie aktar
veri = pd.read_csv("olimpiyatlar_temizlenmis.csv")
veri.head()
```

## Salm Grafii - Kategorik

## Salm Grafii - Dorusal Regresyon

```python
sns.regplot(x='boy', y='kilo', data=veri, marker="+", scatter_kws={'alpha':0.2})
plt.title('Boy ve Kilo Dalm')
plt.show()
```

## Histogram

```python
sns.displot(veri, x = "kilo")
plt.ylabel("Frekans")
plt.title("Kilo Histogram")
plt.show()
```

<a id='9'></a>
## Histogram - Kategorik

```python
sns.displot(veri, x="kilo", hue="cinsiyet")
plt.ylabel("Frekans")
plt.title("Cinsiyete Gre Kilo Histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", col="cinsiyet", multiple="dodge")
plt.show()
```

<a id='10'></a>
## ki Boyutlu Histogram

```python
sns.displot(veri, x="kilo", y="boy", kind="kde")
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("Kilo - Boy Histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy",  hue="cinsiyet", kind="kde")
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("Cinsiyete gre Kilo - Boy Histogram")
plt.show()
```

```python
sns.kdeplot(data = veri, x="kilo", y="boy",  hue="cinsiyet", fill = True)
plt.xlabel("kilo")
plt.ylabel("boy")
plt.title("Cinsiyete gre Kilo - Boy Histogram")
plt.show()
```

<a id='14'></a>
## Keman Grafii

```python
sns.violinplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", data=veri, hue="cinsiyet")
plt.show()
```

```python
sns.violinplot(x="sezon", y="boy", data=veri, hue="cinsiyet", split=True)
plt.show()
```

<a id='15'></a>
## Ortaklk Grafii

```python
sns.jointplot(
    data=veri,
    x="kilo", y="boy", hue="sezon",
    kind="kde")
plt.show()
```

```python
g = sns.JointGrid(data=veri, x="kilo", y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

<a id='16'></a>
## ift/E Grafii

```python
sns.pairplot(veri)
plt.show()
```

```python
g = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```
