# Veri Grselletirme
* **Veri Grselletirme Nedir?**
    * Bugn i dnyasnda veri analizi yoluyla toplanan bu kadar ok bilgi ile, bu verilerin bir resmini izmenin bir yolunu bulmalyz, bylece onlar yorumlayabiliriz. 
    * Veri grselletirme, haritalara veya grafiklere grsel balam salayarak bilginin ne anlama geldiine dair bize net bir fikir verir. 
    * Bu, verileri insan zihninin anlamas iin daha doal hale getirir ve bu nedenle byk veri kmelerindeki eilimleri, modelleri ve aykr deerleri tanmlamay kolaylatrr.
    * Ksaca, veri grselletirme verilerin veya bilgilerin bir grafik, izelge veya dier grsel formatta temsilidir. 
* **Veri Grselletirmenin nemi**
    * Ana ama, veri ve istatistikleri grselletirmek, bilgi edinmek iin ekranlar yorumlamaktr. 
    * Veri grselletirme, veri temizleme, veri yapsn kefetme, aykr deerleri ve olaan d gruplar tespit etme, eilimleri ve kmeleri belirleme, yerel kalplar belirleme, modelleme ktsn deerlendirme ve sonular sunma iin kullanldr.
    * Makine renimi, tahmine dayal analiz gibi analizler yapmay kolaylatrr ve bu analizler, daha sonra sunum iin yararl grselletirmeler ilevi grebilir.
    * Finans, pazarlama, teknoloji, tasarm veya baka herhangi bir alanda alyor olun, verileri grselletirmeniz gerekir. Bu gerek, veri grselletirmenin nemini gstermektedir.
* **Veri Grselletirmede En ok Kullanlan Python Ktphaneleri**
    1. **Matplotlib:** Python temellerini hatrlama blmnde grdmz temel grselletirme ktphenesi.
    1. **Seaborn:** statiksel bir Python veri grselletirme ktphanesidir. (Bu blmde greceiz)
    1. **Plotly:** Etkileimli bir Python veri grselletirme ktphanesidir. (Bu blmde greceiz)



## Seaborn Ktphanesi

* Seaborn kullanclara istatiksel grselletirmeler yapmalar iin yksek seviyeli bir arayz sunar.

    * Matplotlib kiiselletirilebilir fakat plotlar daha anlalr, gze hitap eden ekle getirmek bir hayli zordur.
